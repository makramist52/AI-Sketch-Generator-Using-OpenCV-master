```python
# Description: Sketching the picture using openCV

import  cv2

# get the image location
zayan_img = input("Enter the name of image e.g.(imgname.jgp) : ")

# read image
img = cv2.imread(zayan_img)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=[5,5])
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
```

```python
# convert to grayscaled
grayscaled_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=[5,5])
plt.imshow(grayscaled_img, cmap='gray')
plt.axis("off")
```

```python
# invert the image
inverted_grayscaled_img = 255 - grayscaled_img
plt.figure(figsize=[5,5])
plt.imshow(inverted_grayscaled_img, cmap='gray')
plt.axis("off")
```

```python
# blur the image using gaussian function
blured_img = cv2.GaussianBlur(inverted_grayscaled_img, (21, 21), 0)
plt.figure(figsize=[5,5])
plt.imshow(blured_img, cmap='gray')
plt.axis("off")
```

```python
# invert the blurd image
inverted_blurd_img = 255 - blured_img
plt.figure(figsize=[5,5])
plt.imshow(inverted_blurd_img, cmap='gray')
plt.axis("off")
```

```python
# create the pencil sketch
pencil_blurd_img = cv2.divide(grayscaled_img, inverted_blurd_img, scale = 256.0)
plt.figure(figsize=[5,5])
plt.imshow(pencil_blurd_img, cmap='gray')
plt.axis("off")
```

```python
pencil_blurd_img = cv2.putText(pencil_blurd_img, "Engr. Zia Ur Rehman", (pencil_blurd_img.shape[1]-200, pencil_blurd_img.shape[0]-50), 1, 1, (0, 0, 0), 2)
plt.figure(figsize=[5,5])
plt.imshow(pencil_blurd_img, cmap='gray')
plt.axis("off")
```
